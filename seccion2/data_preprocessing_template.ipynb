# Plantilla de Pre Procesado

# Cómo importar las librerías


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importar el data set


```python
dataset = pd.read_csv('Salary_Data.csv')
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

# Dividir el data set en conjunto de entrenamiento y conjunto de testing


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

# Escalado de variables

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
import sklearn.linear_model as lm
regression = lm.LinearRegression()
# usar matriz de caracteristicas X_train y variable dependiente y_train
regression.fit(X_train, y_train)
```

```python
# Predecir el conjunto de test
y_pred = regression.predict(X_test)
```

```python
# Visualizar los resultados de entrenamiento
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regression.predict(X_train), color = 'blue')
plt.title('Sueldo vs Experiencia (Conjunto de Entrenamiento)')
plt.xlabel('Años de Experiencia')
plt.ylabel('Sueldo (en $)')
plt.show()

# Visualizar los resultados de test
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, regression.predict(X_test), color = 'blue')
plt.title('Sueldo vs Experiencia (Conjunto de Test)')
plt.xlabel('Años de Experiencia')
plt.ylabel('Sueldo (en $)')
plt.show()
```
